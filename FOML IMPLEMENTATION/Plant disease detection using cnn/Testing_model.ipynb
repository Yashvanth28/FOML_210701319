```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

```python
validation_set = tf.keras.utils.image_dataset_from_directory(
    'valid',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(128, 128),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)
class_name = validation_set.class_names
print(class_name)
```

```python
cnn = tf.keras.models.load_model('trained_plant_disease_model.keras')
```

```python
#Test Image Visualization
import cv2
image_path = 'test/Healthy/8ddaa5a5caa5caa8.jpg'
# Reading an image in default mode
img = cv2.imread(image_path)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #Converting BGR to RGB
# Displaying the image 
plt.imshow(img)
plt.title('Test Image')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
image = tf.keras.preprocessing.image.load_img(image_path,target_size=(128,128))
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr = np.array([input_arr])  # Convert single image to a batch.
predictions = cnn.predict(input_arr)
```

```python
print(predictions)
```

```python
result_index = np.argmax(predictions) #Return index of max element
print(result_index)
```

```python
# Displaying the disease prediction
model_prediction = class_name[result_index]
plt.imshow(img)
plt.title(f"Disease Name: {model_prediction}")
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
a=model_prediction
# print(type(a))
print(a)
l=['Powdery', 'Healthy','Rust']
if a in l:
    if a=='Rust':
        image_pat = "Fertilizers/rust_fertilizer.jpg"
        img1 = cv2.imread(image_pat)
        img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
        plt.title(f"Fertilizer Recommendation for: {model_prediction}")# Converting BGR to RGB
        plt.imshow(img2)
    elif a=='Powdery':
        image_pat = "Fertilizers/Late_blight.jpg"
        img1 = cv2.imread(image_pat)
        img2 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
        plt.title(f"Fertilizer Recommendation for: {model_prediction}")# Converting BGR to RGB
        plt.imshow(img2)
    else:
        print("The Potato Leaf is Healthy")
else:
    print("Please select correct image")
```
